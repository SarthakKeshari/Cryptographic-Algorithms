```python
#Step 2 - Convert each letter of the input string into ascii
def letters_to_ascii(s):
    l = []
    for i in s:
        l.append(ord(i))
    
    print("Step 2:",l)
    return l
```

```python
#Step 3 - Writing binary equivalent of ascii and return the combined binary string
def asciiList_to_BinaryString(l):
    s="";
    for i in l:
#         s+=toBinaryString(i)+" "
        s+=toBinaryStringSize8(i)
    print("Step 3:",s)
    return s

def toBinaryStringSize8(num):
    s = ""
    for i in range(8):
        s=str(num%2)+s
        num=num//2
    return s
```

```python
# Step 4 - Append 1 to the result of Step 3
def appending_one(s):
    return s+"1"
```

```python
# Step 5 - Adding zero to the resultant string of step 4 until the string becomes 64 bits less than the multiple of 512
def adding_zeros(s):
    val = len(s)//512
    mul = (val+1)*512
    if((mul-len(s))<64):
        mul+=512
    
    s+=("0"*(mul-len(s)-64))
    
    print("Length after Step 5:",len(s))
    print("Step 5:",s)
    return s
```

```python
# Step 6 - Appending 64 bits representing the length of the original message's binary string
def appending_64bits(s, msgBinaryString):
    msgLen = len(msgBinaryString)
    
    msgLenBinary = ""
    for i in range(64):
        msgLenBinary = str(msgLen%2) + msgLenBinary
        msgLen = msgLen//2
    
    print("Length after Step 6:",len(s+msgLenBinary))
    print("Step 6:",s+msgLenBinary)
    return s+msgLenBinary;
```

```python
# Step 7 - Break the message into blocks of 512 bits
def blockSize512(s):
    l=[]
    while(len(s)>0):
        l.append(s[0:512])
        s=s[512:]
    
    print("Number of message blocks after Step 7:",len(l))
    print("Step 7:",l)
    return l
```

```python
# Step 8 - Divide each block message into 32-bit word
def blockSize32(l):
    l32=[]
    
    for i in l:
        l512 = i
        temp = []
        while(len(l512)>0):
            temp.append(l512[0:32])
            l512=l512[32:]
        l32.append(temp)
    
    for l in l32:
        i = 0
        for a in l:
            print("Word",str(i+1)+":",a)
            i+=1
    
    return l32
```

```python
# Step 9 - Message Schedule for word from index 16 to 63
def toBinaryStringSize32(num):
    s = ""
    for i in range(32):
        s=str(num%2)+s
        num=num//2
    return s

# Step 9.1 - Create a Right Rotate Function
def rightRotate(s, times):
    return s[len(s)-times:]+s[0:len(s)-times]
    
# Step 9.2 - Create a Right Shift Function
def rightShift(s, times):
    return "0"*(times)+s[0:len(s)-times]

# Step 9.3 - Create XOR function
def xor(a,b):
    y = int(a,2) ^ int(b,2)
    return toBinaryStringSize32(y)

# Step 9.5 - Create Add function

def addition(a,b):
#     y = twos_comp_binaryString(int(a, 2), len(a)) + twos_comp_binaryString(int(b, 2), len(b))
    y = (int(a,2) + int(b,2))%(pow(2,32))
    return toBinaryStringSize32(y)

# Step 9.6 - Create a function to apply the following formula - 
# For i from w[16â€¦63]:
# s0 = (w[i-15] rightrotate 7) xor (w[i-15] rightrotate 18) xor (w[i-15] rightshift 3)
# s1 = (w[i- 2] rightrotate 17) xor (w[i- 2] rightrotate 19) xor (w[i- 2] rightshift 10)
# w[i] = w[i-16] + s0 + w[i-7] + s1
def MessageSchedule_16_63(lists):
    rlist = []
    for l in lists:
        i = 16
        for j in range(i,64):
            s0 = xor(xor(rightRotate(l[j-15],7), rightRotate(l[j-15],18)), rightShift(l[j-15],3))
            s1 = xor(xor(rightRotate(l[j-2],17), rightRotate(l[j-2],19)), rightShift(l[j-2],10))
            l.append(addition(addition(l[j-16],s0),addition(l[j-7],s1)))
        rlist.append(l)
    
    for l in rlist:
        i = 0
        for a in l:
            print("Word",str(i+1)+":",a)
            i+=1
    return rlist
```

```python
# Step 13 - Compression using the below methodology
# for i from 0 to 63
# S1 = (e rightrotate 6) xor (e rightrotate 11) xor (e rightrotate 25)
# ch = (e and f) xor ((not e) and g)
# temp1 = h + S1 + ch + k[i] + w[i]
# S0 = (a rightrotate 2) xor (a rightrotate 13) xor (a rightrotate 22)
# maj = (a and b) xor (a and c) xor (b and c)
# temp2 := S0 + maj
# h = g
# g = f
# f = e
# e = d + temp1
# d = c
# c = b
# b = a
# a = temp1 + temp2

def logicalAnd(a,b):
    y = int(a,2) & int(b,2)
    return toBinaryStringSize32(y)

def logicalNot(a):
    y = ""
    for i in a:
        if(i=="0"):
            y+="1"
        else:
            y+="0"
    return y

def compression(lists):
    abcdefghList = []
    for l in lists:
        i = 0
        a = '01101010000010011110011001100111'
        b = '10111011011001111010111010000101'
        c = '00111100011011101111001101110010' 
        d = '10100101010011111111010100111010' 
        e = '01010001000011100101001001111111' 
        f = '10011011000001010110100010001100' 
        g = '00011111100000111101100110101011' 
        h = '01011011111000001100110100011001'
        temp=[]
        for j in range(i,64):
            s1 = xor(xor(rightRotate(e,6), rightRotate(e,11)), rightRotate(e,25))
#             print(s1)
            ch = xor(logicalAnd(e,f),logicalAnd(logicalNot(e),g))
#             print(logicalNot(e))
            temp1 = addition(addition(addition(addition(h,s1),ch),k[j]),l[j])
#             print(temp1)
            s0 = xor(xor(rightRotate(a,2), rightRotate(a,13)), rightRotate(a,22))
#             print(s0)
            maj = xor(xor(logicalAnd(a,b),logicalAnd(a,c)),logicalAnd(b,c))
#             print(maj)
            temp2 = addition(s0,maj)
#             print(temp2)
            h=g
            g=f
            f=e
            e=addition(d,temp1)
            d=c
            c=b
            b=a
            a=addition(temp1,temp2)
        temp.append(a)
        temp.append(b)
        temp.append(c)
        temp.append(d)
        temp.append(e)
        temp.append(f)
        temp.append(g)
        temp.append(h)
        abcdefghList.append(temp)
        
        for abcdefgh in abcdefghList:
            ch = 97
            for i in abcdefgh:
                print(chr(ch),"=",i)
                ch += 1
    return abcdefghList
```
